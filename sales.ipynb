```python


# Group the data by category and product
grouped_data = data.groupby(['Category', 'Product'])

# Calculate the total sales for each category
total_category_sales = data.groupby('Category')['Sales'].transform('sum')

# Calculate the percentage of sales for each product within its category
data['percentage_sales'] = (data['Sales'] / total_category_sales) * 100

data['percentage_sales'] = data['percentage_sales'].round(2)


# Print the results
print(data)
```
